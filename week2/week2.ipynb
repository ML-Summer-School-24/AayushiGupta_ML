```python
#check out converting ordinal data to numeric
#BINNING DATA
def binningFunction(col, cut_points, labels=None):
    minval = col.min()
    maxval = col.max()
    break_points = [minval] + cut_points + [maxval]
    print(break_points)
    if labels is None:
        labels = range(len(cut_points) + 1)
    colBin = pd.cut(col, bins=break_points, labels=labels, include_lowest=True)
    return colBin

# Example cut points and labels for 'pH'
cut_points = [3.3, 3.6]
labels = ["highly acidic", "acidic", "not very acidic"]

# Applying the binning function to the 'pH' column
orgds["Acidity based on pH"] = binningFunction(orgds["pH"], cut_points, labels)


orgds
```
